```python
import csv
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans
%matplotlib inline
df = pd.read_csv('tmp.txt', sep = '|', low_memory = False)
df
```

```python
cluster_num = 5
kmeans = KMeans(n_clusters = cluster_num)
kmeans.fit(df.ix[:,['actcount','med_ml_num','buynum', 'grind','quests_speed']])
best = kmeans
for x in range(5):
    kmeans = KMeans(n_clusters = cluster_num)
    kmeans.fit(df.ix[:,['actcount','med_ml_num','buynum', 'grind','quests_speed']])
    if kmeans.inertia_<best.inertia_:
        best = kmeans
kmeans = best
print "Inertia "+"{:10.4f}".format(kmeans.inertia_)
for x in kmeans.cluster_centers_:
    print '  '.join("{:10.4f}".format(i) for i in x)
with open('cluster_centers.txt','w') as f:
    writer = csv.writer(f, delimiter='|',
                            quotechar='"', quoting=csv.QUOTE_MINIMAL)
    writer.writerow(['actcount','med_ml_num','buynum', 'grind','quests_speed'])
    for x in kmeans.cluster_centers_:
        writer.writerow(x)
    f.close()
```

```python
prediction = kmeans.predict(df.ix[:,['actcount','med_ml_num','buynum', 'grind','quests_speed']])
with open('clusters.csv','w') as f:
    writer = csv.writer(f, delimiter='|',
                            quotechar='"', quoting=csv.QUOTE_MINIMAL)
    writer.writerow(['id','cluster_index'])
    for i in xrange(len(prediction)):
        writer.writerow([df['id'].ix[i],prediction[i]])
    f.close()
```
