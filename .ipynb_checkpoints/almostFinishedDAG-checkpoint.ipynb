```python
import networkx as nx
import xml
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import matplotlib.gridspec as gridspec
import matplotlib.font_manager as font_manager
import xml.etree.ElementTree as ET

filename = 'ghost_dag.txt'

try:
    tree = ET.parse(filename)
except:
    s = '1'
    lines = ''
    with open(filename, 'r') as f:
        while s:
            s = f.readline()
            sp = ''.join(s.split())
            if (sp[:2] == "<#") or (sp[:3] == "</#"):
                print sp
            else:
                lines = lines + s + '\n'          
    with open('fine_' + filename, 'w') as f:
        f.write(lines)
    tree = ET.parse('fine_' + filename)
    
jobtree = tree.getroot()
jobs = jobtree.getchildren()[0].getchildren()
G = nx.DiGraph()
for job in jobs:
    jobname = job.attrib['name']
    jobtype = job.attrib['jobType']    
    param_num = 0
    
    children = job.getchildren()
    for child in children:
        if child.tag == 'parameters':
            param_num = len(child._children)
    
    print jobname, jobtype, param_num
    G.add_node(jobname, jobtype = jobtype, param_num = param_num)

for job in jobs:
    jobname = job.attrib['name']
    children = job.getchildren()
    for child in children:
        if child.tag == 'dependencies':
            dependencies = child._children
            for dependency in dependencies:
                dep = dependency.attrib['name']
                G.add_edge(dep, jobname)
color_dict = {'CLUSTER_ACTIVATOR': 'g', 
              'DOCKER_TASK': 'r', 
              'PYTHON_SCRIPT': 'b',
              'HIVE_QUERY': 'c', 
              'root' : 'k'}


labels = {}
print "Graph description"
for node in G.node:
    print node, G.node[node]
    labels[node] = node + '(' + str(G.node[node]['param_num']) + ')'
```

```python
cnt = 0
lb2 = labels
for key in labels.keys():
    cnt += 1
    lb2[key] = cnt
inv_lb2 = {v: k for k, v in lb2.items()}
import pydotplus
from networkx.drawing.nx_pydot import graphviz_layout

%matplotlib inline
pos = graphviz_layout(G, prog='dot', args='')
fig = plt.figure(figsize=(30, 15))
plt.suptitle('"'+ jobtree.attrib['name']+'" analysis graph', fontsize = 18)

# sizes of subplots
grid_spec = gridspec.GridSpec(1,3, width_ratios=[3, 1, 1])

# nodes of the graph
ax = fig.add_subplot(grid_spec[0])
# nx.draw(G, pos, node_size=400, alpha=0.5, node_color=clrs, labels = lb2, arrows=False, width = 2)
nx.draw_networkx_edges(G, pos, node_size=400, alpha=0.5, labels = lb2, arrows=False, width = 1.3)
# draw subgraph for each type
for t in color_dict:
    nx.draw_networkx_nodes(G, pos, nodelist = filter(lambda x:G.node[x]['jobtype']==t, G.node),
                           node_size=400, alpha=0.5, node_color=color_dict[t], labels = lb2)
nx.draw_networkx_labels(G, pos, node_size=400, alpha=0.5, labels = lb2, arrows=False, width = .3)
ax.axis('off')

# legend of the graph
vertical_margin_multiplier = 1.3
font_size = 12
font_properties = font_manager.FontProperties(size = font_size)
fheight = font_properties.get_size_in_points()
tick = 0.4/(len(inv_lb2)*vertical_margin_multiplier)
tick = (1+vertical_margin_multiplier)*fheight if (1+vertical_margin_multiplier)*fheight < tick else tick

# nodes legend
ax = fig.add_subplot(grid_spec[1])
cntx = 0.95


for key in xrange(len(inv_lb2)):
    cntx -= tick
    ax.text(0, cntx,str(key+1)  + (' 'if key<9 else '') + ' : ' + str(inv_lb2[key+1]), fontsize=font_size)
ax.axis('off')

# drawing of color legend
ax = fig.add_subplot(grid_spec[2])
ax.axis('off')
cntx = 0.95
rect_size = tick/2.3
margin = 0.001
for key in color_dict:
    cntx -= tick
    ax.add_patch(
        patches.Rectangle(
            (0, cntx),   # (x,y)
            rect_size,          # width
            rect_size,          # height
            fill = True,
            color = color_dict[key]
        )
    )
    ax.text(rect_size + margin, cntx, ' : '+key)
plt.savefig("out.png")
```
